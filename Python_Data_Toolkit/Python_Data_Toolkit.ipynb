**Ques 5.** What are the differences between NumPy arrays and Python lists?

**Ans.** Here are some key differences between NumPy arrays and Python lists:

- **Performance:** NumPy arrays are generally much faster for numerical operations due to their fixed type and C-based implementation. Python lists are slower because they can hold elements of different data types.

- **Memory Usage:** NumPy arrays are more memory-efficient as they store elements of the same type in a contiguous block of memory. Python lists use more memory due to storing pointers to objects.

- **Functionality:** NumPy arrays support a wide range of mathematical functions and vectorized operations, which are not directly available for Python lists.

- **Data Types:** NumPy arrays store elements of a single data type, while Python lists can store elements of different data types.
Creating: NumPy arrays are created using the `numpy.array()` function, while Python lists are created using square brackets `[]`.

**Ques 7.** What does the term “vectorized operation” mean in NumPy?

**Ans.** In NumPy, a "vectorized operation" refers to performing operations on entire arrays at once, rather than iterating through individual elements using explicit loops. This is possible because NumPy operations are implemented in C, which makes them significantly faster and more efficient for numerical computations compared to standard Python loops. It's essentially applying a function to all elements of an array (or between arrays) simultaneously.

**Ques 16.** Explain the difference between apply() and map() in Pandas?

**Ans.**  The `apply()` and `map()` methods in Pandas are used for applying functions to data, but they differ in their usage and the types of data they operate on:

- **`map()`:** This method is used to substitute each value in a **Series** with another value. It takes a dictionary, a Series, or a function as an argument. It's primarily used for element-wise transformation of values in a Series.
- **`apply()`:** This method is more versatile and can be used on both **Series and DataFrames**. When used on a Series, it applies a function to each element. When used on a DataFrame, it can apply a function along either axis (rows or columns). It's more general and can handle more complex operations than `map()`.

In summary, `map()` is specifically for element-wise operations on a Series, while `apply()` is more flexible and can be used for element-wise or axis-wise operations on both Series and DataFrames.

- **Fast Fourier Transform (FFT):** NumPy includes functions for performing FFT and inverse FFT, useful in signal processing.
- **Masked Arrays:** These are arrays that can have missing or invalid entries. NumPy's `ma` module provides tools for working with masked arrays.
- **Memory Mapping:** NumPy allows memory-mapping of arrays stored in files, enabling processing of very large datasets that don't fit into memory.

**Ques 20.** Why is NumPy's array slicing faster than Python's list slicing?

**Ans.** NumPy's array slicing is faster than Python's list slicing due to several factors:

- **Homogeneous Data Type:** NumPy arrays store elements of a single, consistent data type. This allows NumPy to store the elements in a contiguous block of memory.
- **Contiguous Memory Allocation:** Because elements are stored contiguously, accessing elements via slicing in NumPy is a simple matter of calculating memory offsets. This is very efficient.
- **C Implementation:** NumPy's core operations, including slicing, are implemented in C. This compiled code is significantly faster than Python's interpreted code.
- **No Overhead for Type Checking:** Since all elements are of the same type, NumPy doesn't need to perform type checking for each element during slicing, unlike Python lists which can hold elements of different types.
- **Optimized Algorithms:** NumPy is specifically designed for numerical operations and its slicing implementation is highly optimized for this purpose.

In contrast, Python lists can store elements of different data types, which means they don't necessarily store elements contiguously in memory. Slicing a Python list often involves creating a new list and copying elements, which is a less efficient process than NumPy's memory offsetting.

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

# import library
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [10, 20, 30, 40, 50],
    'D': [100,200,300,400,500]
}

# Data Convert to DataFrame
df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# 8. Create a line plot with multiple lines using Matplotlib Generate a Pandas.

# import Libriry
import pandas as pd
import matplotlib.pyplot as plt

# Creating Data

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales_A': [100, 120, 150, 170, 160, 180],
    'Sales_B': [90, 110, 130, 140, 150, 170],
    'Sales_C': [80, 95, 120, 135, 145, 160]
}

# Data Convert to DataFrame
df = pd.DataFrame(data)

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot multiple lines
plt.plot(df['Month'], df['Sales_A'], label='Sales_A', marker='o', linestyle='-')
plt.plot(df['Month'], df['Sales_B'], label='Sales_B', marker='x', linestyle='--')
plt.plot(df['Month'], df['Sales_C'], label='Sales_C', marker='s', linestyle='-.')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison')

# Add a legend
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
