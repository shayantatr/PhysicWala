**Question 1:**  What is a random variable in probability theory?  

Answer:  A **random variable** is a numerical value that represents the outcome of a random event. It assigns a number to each possible outcome in a **sample space**.

**Example:**
If you toss a coin:
- Let X = number of heads.

  Possible outcomes: {0, 1}

  Here, X is a random variable.

**Types of random variables:**

1. Discrete Random Variable: Takes on a countable number of values (e.g., 0, 1, 2, 3...). Example: Number of heads in 5 coin tosses

2. Continuous Random Variable: Takes on an uncountable number of values (any value within a range). Example: Height, temperature, or time.

---

**Question 2:** What are the types of random variables?

**Answer:** There are two main types of random variables:

1.  **Discrete Random Variable:** This type of random variable can only take on a finite number of distinct values or an infinite sequence of countable values. Examples include the number of heads in a series of coin flips, the number of cars passing a point in an hour, or the number of defects in a manufactured product.

2.  **Continuous Random Variable:** This type of random variable can take on any value within a given range or interval. Examples include the height of a person, the temperature of a room, the weight of an object, or the time it takes to complete a task.
---

**Question 3:** Explain the difference between discrete and continuous distributions.  

**Answer:** The difference between discrete and continuous distributions can be summarized in the following table:

| Feature        | Discrete Distribution     | Continuous Distribution        |
| :---------------- | :------------------------------- | :---------------------------|
| **Random Variable** | Discrete Random Variable     | Continuous Random Variable                  |
| **Possible Values** | Finite or countably infinite distinct values | Any value within a given range or interval  |
| **Probability**   | Assigned to individual values            | Assigned to a range of values (area under curve) |
| **Examples**      | Binomial, Poisson                        | Normal, Exponential  |

**Example:**
- Discrete: Probability of getting 3 heads in 5 tosses (Binomial).
- Continuous: Probability that temperature is between 20°C and 25°C (Normal).

**Question 4:** What is a binomial distribution, and how is it used in probability?

**Answer:** A **Binomial Distribution** models the number of successes in a fixed number of **independent Bernoulli trials**, each having the same probability of success.

**Formula:**

$P(X=k) = \binom{n}{k} * p^k * (1-p)^{n-k}$

**Where:**
- $n$ = number of trials
- $k$ = number of successes
- $p$ = probability of success
- $(1-p)$ = probability of failure


**Example:**

If you flip a coin 10 times (n=10, p=0.5), the probability of getting exactly 6 heads follows a **Binomial Distribution**.

**Used for:**
- Quality control
- Coin toss problems
- Success/failure experiments (like pass/fail, yes/no outcomes)
---

**Question 5:** What is the standard normal distribution, and why is it important?

**Answer:** The **standard normal distribution**, also known as the Z-distribution, is a special case of the normal distribution. It has a mean of 0 and a standard deviation of 1.

It's important for several reasons:

* **Standardization:** Any normal distribution can be converted into a standard normal distribution by standardizing the values. This allows us to compare values from different normal distributions.
* **Probability Calculation:** We can use the standard normal distribution table (Z-table) to find the probability of a value falling within a certain range, which simplifies calculations for any normal distribution after standardization.
* **Foundation for Statistical Inference:** Many statistical tests and confidence intervals are based on the properties of the standard normal distribution.
---

**Question 6:** What is the Central Limit Theorem (CLT), and why is it critical in statistics?

**Answer:** The **Central Limit Theorem** states that:

>The sampling distribution of the sample mean approaches a **Normal distribution** as the sample size (n) increases — regardless of the population's original distribution — provided n is large enough (usually n ≥ 30).

**Why it's important:**

* Allows the use of **Normal distribution** in many real-world problems.
* Makes inference (like confidence intervals and hypothesis tests) possible even when the population distribution is unknown
---

**Question 7:**  What is the significance of confidence intervals in statistical analysis?

**Answer:** A **Confidence Interval (CI)** gives a range of values that likely contains the true population parameter (like mean or proportion) with a certain confidence level (e.g., 95%).

**Example:**
* A 95% confidence interval for the mean = [45, 55]
* → We are 95% confident that the true mean lies between 45 and 55.

**Significance:**
* Shows uncertainty in estimation.
* Helps interpret precision of sample estimates.
* Widely used in research, surveys, and experiments.
---

**Question 8:** What is the concept of expected value in a probability distribution?  

**Answer:** The **Expected Value (E[X])** is the **average or mean** value of a random variable — the long-term average outcome if an experiment is repeated many times.

**For Discrete Variables:**

$E[X]=\sum_{}x_iP(x_i)$

**For Continuous Variables:**

$E[X]=\int_{}xf(x)dx$

**Example:**
If you roll a fair 6-sided die:
- $E[X]$ = (1+2+3+4+5+6)/6 = 3.5

Meaning:
- If you roll the die many times, the average outcome approaches 3.5.

```python
'''
Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and
standard deviation = 5. Compute its mean and standard deviation using NumPy,
and draw a histogram to visualize the distribution.
'''

# Step 0: Import Library
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Step 2: Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print(f"Mean of the data: {mean:.2f}")
print(f"Standard Deviation of the data: {std_dev:.2f}")

# Step 3: Draw a histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Normally Distributed Data (μ=50, σ=5)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add a vertical line for the mean
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {mean:.2f}')
plt.legend()
plt.show()
```

```python
'''
Question 10:  You are working as a data analyst for a retail company.
The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

-> Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
-> Write the Python code to compute the mean sales and its confidence interval.
'''

# Import Liabrary
import numpy as np
import scipy.stats as stats

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Step 1: Compute basic statistics
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)  # sample standard deviation
n = len(daily_sales)

# Step 2: Compute the standard error
std_error = std_sales / np.sqrt(n)

# Step 3: Compute the t-critical value for 95% confidence interval
confidence_level = 0.95
t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

# Step 4: Compute margin of error
margin_of_error = t_critical * std_error

# Step 5: Compute confidence interval
lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

# Step 6: Display results
print(f"Sample Mean Sales: {mean_sales:.2f}")
print(f"Standard Deviation: {std_sales:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
